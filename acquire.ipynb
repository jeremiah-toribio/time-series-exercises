# Dictionary Breakdown
['count'] = total number of values\
['previous'] = last page of stored values\
['next'] = next page of stored values\
['results'] = actual data dictionary stored

```python
num_people = people['count']
next_page = people['next']
previous_page = people['previous']

print(f'Number of people: {num_people}')
print(f'Next Page: {next_page}')
print(f'Previous Page: {previous_page}')
```

```python
num_results = len(people['results'])
max_page = math.ceil(num_people / num_results)

print(f'Number of results on page: {num_results}')
print(f'Max Num of Pages: {max_page}')
```

```python
def swapi():
    data = pd.DataFrame()
    objects = ['people','planets','starships']
    for obj in objects:
        url_response = requests.get(f'https://swapi.dev/api/{obj}')
        url = url_response.json()
        # count number of observations in total
        num_count = url['count']
        # num. results on page
        num_results = len(url['results'])
        # defining outer range
        max_page = math.ceil(num_count / num_results)
        for page in range(1,max_page+1):
            url_response = requests.get(f'https://swapi.dev/api/{obj}/?page={page}')
            url = url_response.json()
            results = url['results']
            df = pd.DataFrame(results)
            data = pd.concat([data, df], ignore_index=True)

    data = data.drop_duplicates(subset=['name']).reset_index(drop=True)

    return data

df = swapi()
```
